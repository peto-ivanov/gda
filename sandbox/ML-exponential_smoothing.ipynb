# Exponential Smoothing

[Wiki Article (Exponential smoothing)](https://en.wikipedia.org/wiki/Exponential_smoothing)

The raw data sequence is often represented by $\{x_0\}$ beginning at time $t=0$, and the output of the exponential smoothing algorithm is commonly written as $\{s_t\}$, which may be regarded as a best estimate of what the next value of $x$ will be. When the sequence of observations begins at time $t=0$, the simplest form of exponential smoothing is given by the formulas:

$$s_0=x_0$$  
$$s_t=\alpha x_{t}+(1-\alpha )s_{t-1}$$

where $\alpha$  is the smoothing factor, and $0<\alpha<1$.

```
import numpy as np
import matplotlib.pyplot as plt
```

For demonstration purposes, let's create random series:

```
# Initialize RNG
rng = np.random.default_rng()

# x = linear space in <0, 100>
# y = 100 random nubers in <0, 10>
x = np.linspace(0, 1000, num=1000)
y = rng.random(1000, dtype=np.float32)*10
```

Function to compute exponential smoothing of an iterable with given $\alpha$.

```
def exp_smooth(a, alpha=1):
    """ Compute exponential smoothing of iterable and return a numpy array."""
    x = list(a)
    s = []
    s.append(x[0])
    
    for t, _ in enumerate(x):
        if t != 0:
            s.append(alpha*x[t] + (1-alpha)*s[t-1])
    
    return np.asarray(s, dtype=np.float32)
```

```
# Docstring becomes __doc__ attribute of the Python object
print(exp_smooth.__doc__)
```

Let's plot exponential smoothings using different $\alpha$:

```
y2 = exp_smooth(y, 0.5)
y3 = exp_smooth(y, 0.3)
y4 = exp_smooth(y, 0.1)
y5 = exp_smooth(y, 0.05)
```

```
fig, ax = plt.subplots(figsize=(18,6))

plt.plot(x, y)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
plt.show()
```
