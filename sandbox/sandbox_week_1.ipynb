```python
dict = {
    "apple": [3, 5],
    "pear": [4, 6]
}

# Vytvorenie DataFramu zo slovníka
df=pd.DataFrame.from_dict(dict, orient='columns')

df
```

```python
df.loc[len(df.index)] = [5, 7]
```

```python
print(len(df.index))

# 1 parameter -> locate index (row)
print(df.loc[2])

# 2 parametre -> locate index + column = jedna hodnota
print(df.loc[2, "apple"])

# 2 parametre (zoznam indexov + column) = hodnota v stlpci pre riadky zo zoznamu
print(df.loc[[1, 2], "pear"])

df
```
