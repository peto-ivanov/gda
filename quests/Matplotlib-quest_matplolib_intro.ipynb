```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt

link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link).iloc[:,[0, 1, 2, -3]]

df_weather['DATE'] = pd.to_datetime(df_weather['DATE'])
```

```
# Complete the code below :

# Initializing figure and axes
fig, ax = plt.subplots()

# Initialize ax1
ax1 = plt.subplot(111)

ax1.scatter(x = df_weather["DATE"], y = df_weather["MAX_TEMPERATURE_C"], c= 'red', label="Max")
ax1.scatter(x = df_weather["DATE"], y = df_weather["MIN_TEMPERATURE_C"], c= 'blue', label="Min") #You have to complete this code. You could probably try to copy the previous line, and change max temperature by min temperature ?

# Title
ax1.set_title("MIN and MAX Temperatures")

# Legend
ax1.legend(title="Temperatures", loc="upper left")

plt.show()
```

## 2 charts on the same figure
Now, complete this code to have MAX and MIN Temperature in 2 charts, MAX on the left side and MIN on the right side 

The X axis have to be the date of the year

Each graph must have a title, a legend for the colors, and of course, readable axis (ticks)

Here is an example of what is expected :


Capture d’écran 2021-08-26 à 17.27.31.png

```
# Complete the code below :

# Initializing figure and axes
fig, ax = plt.subplots(figsize = (15,5))

# Initialize ax1
ax1 = plt.subplot(121)
ax1.scatter(x = df_weather["DATE"], y = df_weather["MAX_TEMPERATURE_C"], c= 'red', label="Max")
ax1.set_yticks(range(-5,45,5))
ax1.set_title("Temperature MAX")
ax1.set_ylabel('Temperature MAX')
ax1.legend()

# Initialize ax2
ax2 = plt.subplot(122)
ax2.scatter(x = df_weather["DATE"], y = df_weather["MIN_TEMPERATURE_C"], c= 'blue', label="Min")
ax2.set_yticks(range(-5,45,5))
ax2.set_title("Temperature MIN")
ax2.set_ylabel("Temperature MIN")
ax2.legend()

plt.show()
```
