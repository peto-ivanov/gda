# Mission 1

Display the following information [for this dataset](https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user).
- The last 5 rows
- The first 5 rows
- The type of each column

Note : start coding something by typing `pd.read...` and explore the different options available, especially the `sep` parameter.


```
import pandas as pd

# Read the DataFrame from the URL
# I gave the columns some names
df = pd.read_csv(
    "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/u.user",
    sep="|",
    header=None,
    names=["age", "sex", "job", "id"])

# The last 5 rows
print("The last 5 rows:")
print(df.iloc[-5:,:])

# The first 5 rows
print("The first 5 rows:")
print(df.iloc[:5,:])

# The type of each column
print("Column types:")
print(df.dtypes)

```

# Mission 5

Rename the columns [of this dataset](https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv), try using the following 2 different methods to achieve this:
- Setting column names directly when reading the file 
- Editing after creating the dataframe ([This resource could be useful](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html))

```
import pandas as pd

# Setting column names directly when reading the file
df = pd.read_csv(
    "https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv",
    sep=",",
    header=0,
    names=["City", "Colors", "Shape", "State", "Datetime"]
)

print(df.head(10))

# Editing after creating the dataframe
df.rename(
    columns={
      "City": "Town",
      "Colors": "Colored",
      "Shape": "Shaped",
      "State": "Area",
      "Datetime":"Date and Time"},
    inplace=True
  )

print(df.head(10))
```
