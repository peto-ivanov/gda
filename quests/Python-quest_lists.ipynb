# Mission 1

Create a script that takes a list and displays only the elements (and their position) that have positions that are even (0, 2, 4, 6...).

Example list = ["a", "b", "c", "d", "e"], the script should return : 
- a position 0
- c position 2
- e position 4

```
# Let's create the example list
example = ["a", "b", "c", "d", "e"]

# Let's try to do this using enumerate and list comprehension
print("Using list comprehension:")
[print(f"{value} position {position}") for position, value in enumerate(example) if position % 2 == 0]

# Let's try this in "no comprende listas" mode
print("Traditional approach:")
for position, value in enumerate(example):
  if position % 2 == 0:
    print(f"{value} position {position}")
```

# Mission 5

Create a script that does the same thing as the pop method for a list containing only characters. The difference is that this function displays a message indicating the number of characters that are identical to the deleted character and still remain in the list. In other words, if there were 3 "b" in the list, and one has just been deleted by your function, the script will display "there are still 2 letters b".

```
# Set up example list
my_list = [character for character in "the number of characters"]

# Define my new pop function
def mypop(a):

  # Get the last element value
  letter = a[-1]

  # Now delete it
  del(a[-1:])

  # How much of these deleted charaters remain?
  count = a.count(letter)

  # Result
  print(f"there are still {count} letters {letter}")
  return


# How it looks in the beginning
print(my_list)

# Pop 5 times
for i in range(5):
  mypop(my_list)

# How it looks after
print(my_list)
```

# Mission 6

Create a script with 3 lists, and indicates the two that are "closest". That is to say, the two that have the most elements in common.

```
from random import randint
from functools import reduce

# How many lists do we want to compare?
n = 3

# How many elements per list?
m = 10

# What would be the range of random numbers in elements?
p = 6

# Let's make up the lists
lists = [[randint(0,p) for i in range(m)] for list_index in range(n)]

# Let's print them out
[print(f"List {i} = {lists[i]}") for i in range(n)]

# Definition of my comparison function (simple)
def compare(a,b):
  if a == b:
    return 1
  else:
    return 0

# Define maximum similarity result
max_result = 0

# Compare all doublets of lists
for list_a in range(n):
  for list_b in range(n):

    # We only compare upwards (1 with 2), i.e. we don't compare 2 with 1 (backwards)
    if list_a < list_b:

      # Using map() to compare individual elements
      # + using reduce() to count similar elements
      result = reduce(lambda a,b:a+b, map(compare, lists[list_a], lists[list_b]))

      # Some user-friendly output, so we can actually understand what's going on
      print(f"Comparing lists {list_a} and {list_b}.")
      print(list(map(compare, lists[list_a], lists[list_b])))
      print(f"Similarity result: {result}")

      # This is what we do if we find next better result
      if result > max_result:
        max_result = result
        similar_lists = [[list_a, list_b]]

      # This is what we do if we find another two lists with the same good score
      elif result == max_result:
        similar_lists.append([list_a, list_b])

# And final result is written here
print(f"Most similar lists are {similar_lists} with similarity score of {max_result}")
```

# Mission 7

Create a script that displays all the items in a list of int from smallest to the largest. In other words, a script that sorts the list from smallest to largest.

```
from random import randint

# Let's make random list
# How many elements per list?
m = 10

# What would be the range of random numbers in elements?
p = 6

# Let's make up the list
my_list = [randint(0,p) for i in range(m)]

# Before
print(my_list)

# Sorting
my_list.sort()

# After
print(my_list)
```
