# Mission 4

Take the dictionary generated from mission 3. Delete any key value pair randomly, except for total_campus.

```
from random import randint

# Recreating the dictionary just in case something bad happened before
wildcodeschool = {'business intelligence analyst': 5,
 'data_analyst': 8,
 'data_scientist': 2,
 'dev_web': 13,
 'total_campus': 19}

print(wildcodeschool)

# Let's make a random integer. We pick in interval <1, dictionary size - 1>
# so we already think about not touching "total_campus" key
# Example = dictionary is 10 items long, we pick number from <1,9>
random_int = randint(1,len(wildcodeschool)-1)

# Debug
print(f"Random integer = {random_int}")

# This is the magic
# we .pop from dictionary using key, which is picked from a list
# That list is created from keys of original dictionary EXCLUDING 'total_campus'
# And on top of that we only pick the key, which is numbered with the random integer.
[wildcodeschool.pop(key) for index, key in enumerate([key for key in wildcodeschool.keys() if key != 'total_campus']) if index == (random_int - 1)]

# Result
print(wildcodeschool)

```
