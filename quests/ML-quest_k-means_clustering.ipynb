# K-Means Clustering

Practice by doing a __clustering__ using the __dataset Iris__.

* In scikit-learn, import __the Iris dataset__, it is a great classic of data. It contains 3 species of flowers (irises) according to their length and width of petal and sepal.

```python
# Importing base packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

sns.set_style("darkgrid")
```

```python
# Loading iris data
(X, y) = load_iris(return_X_y=True, as_frame=True)
```

We have now _Iris_ data sitting inside _X_ dataframe.

* Train your model to find 3 clusters, from the 4 numeric columns.

```python
# Inspecting the _X_:
sns.pairplot(data=X)
```

```python
# Importing K-Means
from sklearn.cluster import KMeans

# Stupid way to save efforts.
# We run KMeans with parameter n (number of clusters), then we visualize the results
# as requested in the quest (comparison with real "species" samples)
def do_stuff(n):
    # Initiating KM_model with 3 clusters
    KM_model = KMeans(n_clusters=n)

    # Loading iris data
    (X, y) = load_iris(return_X_y=True, as_frame=True)

    # Fitting the clustering model
    KM_clusters = KM_model.fit_predict(X)

    # Doing the graphs
    fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(16,6))
    sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue=KM_clusters, data=X, palette="Set2", ax=axs[0])
    sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue=y, data=X, palette="Set2", ax=axs[1])
    sns.scatterplot(x=KM_model.cluster_centers_[:,2], y=KM_model.cluster_centers_[:,3], s=100, marker="+", color="black", label="centroid", ax=axs[0])
    plt.show()

do_stuff(3)
```

* Draw another model with 2 clusters, then displays the same graphs

```python
do_stuff(2)
```

* Same for 4 clusters

```python
do_stuff(4)
```
