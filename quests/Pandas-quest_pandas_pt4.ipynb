Still using this dataset: http://bit.ly/imdbratings
- Which actor is mentioned the most often?


```
import pandas as pd

# Read CSV via URL
df = pd.read_csv("http://bit.ly/imdbratings")

# This is the whole magic:
(df.loc[:, "actors_list"]
 .apply(eval)
 .explode(ignore_index=True)
 .value_counts()
 .head(1)
)
```
