Create a "fruit" dataframe that reflects the above example **using a list** data structure

[This resource could help you.](https://www.geeksforgeeks.org/create-a-pandas-dataframe-from-lists/)

```
import pandas as pd

# Create list - must be 1 row (sublist) in list
lst = [[30, 21]]

# We have two series, so here are the series names
col = ["Apples", "Bananas"]

# Debug
print(lst)

# Make the DataFrame
fruit = pd.DataFrame(lst, columns=col)

print(fruit)
```

# Mission 2

Create a "fruit_vendors" dataframe that reflects the below example **using a dictionary** data structure

![](https://i.imgur.com/CHPn7ZF.png)

[This resource could help you.](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.from_dict.html)

```
# Create dictionary
# This one is index shaped {index: {column values}, etc.}
dict = {
    "2017 Sales": {
        "Apples": 35,
        "Bananas": 21},
    "2018 Sales": {
        "Apples": 41,
        "Bananas": 34}
}

# Create DataFrame (had to use orient='index', so it's done properly)
fruit_vendors = pd.DataFrame.from_dict(dict, orient='index')

print(fruit_vendors)
    
```

# Mission 3

Use a loop to display the values from each column in the dataframe from mission 2.


```
# Printing values from all columns from entire dataframe using loops
[[print(value) for col, value in enumerate(fruit_vendors[rows])] for rows in fruit_vendors]

# Or there is nicer way printing out entire column contents
[print(fruit_vendors[col]) for col in fruit_vendors.columns]
```
