# Mission 3

Answer the question below:
- What is the median star rating? [This ressource could be useful](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.median.html)
- What is the genre of the most highly rated film? [This ressource could be useful](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.max.html)
- How many (in percentage) of films belong to each genre? [This ressource could be useful](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html)

```
# Check for the median in "star_rating" Series
df["star_rating"].median()
```

```
# What is the genre of the most highly rated film?
# First we find the label of the index with maximum in the "star_rating" Series:
# df["star_rating"].idxmax()
# Then we use it in locating the record and genre
df.loc[df["star_rating"].idxmax(), "genre"]
```

```
# How many (in percentage) of films belong to each genre?

# Using .value_counts with normalize=True
# in the end multiply by 100 to get "percents"
df["genre"].value_counts(normalize=True)*100
```

# Mission 4

Display only the rows with films that belong to the lowest rated genre (based on the average rating of the films in this genre)
[This resource could be useful](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.core.groupby.GroupBy.mean.html)

```
# Create another DataFrame which has means of movie ratings grouped by "genre"
meandf = df.groupby("genre").mean()

# Get the genre with the minimum mean rating
# and use the genre name as a filter.
df[ df["genre"] == meandf["star_rating"].idxmin("star_rating")]
```

# Mission 5

Display films which are longer than 200 minutes **and** which belong to a genre with a name containing the letter 'h' (either lower case or upper case)


```
# Using .loc with filtering function combining
# df["duration"] > 200 
# and
# df["genre"].str.contains("h")
df.loc[(df["duration"] > 200) & (df["genre"].str.contains("h"))]
```

# Mission 6

Which **actor trio** is mentioned most often in the entire dataframe?

```
# Assuming the task is to get TOP 3 actors mentioned in the DF movies
import pandas as pd

# We need this literal_eval function to do data cleaning
from ast import literal_eval

# Read CSV via URL
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv")

# Data cleaning in actors_list
actor_series = (df.loc[:, "actors_list"]
                  .str.replace(r'u\'', '\'')
                  .str.replace(r'u\"', '\"')
                  .apply(literal_eval)
                  .explode(ignore_index=True)
                )

# Convert back to DataFrame, but add counts
df = pd.DataFrame(actors_series.value_counts())

# Display top 3 actors
df.head(3)

# This one was SUPER HARD to do
```
