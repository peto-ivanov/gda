# Mission 3

Create a script that only displays a particular part of a string it receives. In other words, make it so that you are able to decide the exact character range (e.g. 2nd to 5th character) to be displayed.

```
# Let's get a string
my_string = input("Write a string to inspect: ")

# Let's initiate start and stop positions
# ... but as strings so the while loops begin
start = ''
stop = ''

# Let's ask user for the new values
while not isinstance(start, int):
  start = input("Enter starting position (integer): ")
  
  # Compensate the python internal positioning
  # (human would ask for 1 where Python undersands 0)
  start = int(start) - 1

while not isinstance(stop, int):
  stop = input("Enter ending position (integer): ")

  # This time no compensation is necessary.
  # (human would ask for 5 where for Python this is 6th character, which is
  # right excluded limit, so the 5th character is the last one)
  stop = int(stop)

print(my_string[start:stop])
```

# Mission 4

Find 3 ways to return the following string without including the spaces at the beginning and the end: " Data Analyst ".

```
# Initiate the string
sentence = " Data Analyst "

# First method
# By printing a substring
print(sentence[1:-1])

# Second method
# By using method .strip()
print(sentence.strip())

# Thirs method
# By using string.split() and string.join() method to create new string using
# "space" as separator. This extra spaces will not be taken.
print(' '.join(sentence.split()))
```

# Mission 5

Create a script that converts all vowels to uppercase and all consonants to lowercase.

```
# Make some lists we need (only lowercase, see below)
vowels = ['a', 'e', 'i', 'o', 'u', 'y']
consonants = ['b', 'c', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'm', 
              'n', 'p', 'q', 'r', 's', 't', 'v', 'w', 'x', 'z']

# Get a string
sentence = input("Write a string to \"convert\":")

# Make an empty "result"
result = ''

# Scan for characters
for i in range(len(sentence)):
  
  # We convert all characters to lowercase, so we don't need lengthy lists
  # to check against:
  if sentence[i].lower() in vowels:
    result = result + sentence[i].upper()
  elif sentence[i].lower() in consonants:
    result = result + sentence[i].lower()
  else:
    result = result + sentence[i]

# Done, print.
print(result)
```
