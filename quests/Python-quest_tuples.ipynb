
# Mission 4

Create 2 tuples that contain only integers.
Create a script that compares them to each other and indicates which one of the tuples returns the highest total value after adding all their values together. 

```
from random import randint
from functools import reduce

# Creating two tuples
t1 = tuple([randint(1,10) for i in range(10)])
t2 = tuple([randint(1,10) for i in range(10)])

# Function to compute the total value of the integer tuple
def totval(t):
  return reduce(lambda a,b:a+b, t)

# Debug
print(f"Tuple 1 = {t1} [Total value = {totval(t1)}]")
print(f"Tuple 2 = {t2} [Total value = {totval(t2)}]")

# Find the higher total value
if totval(t1) > totval(t2):
  print("Tuple 1 total value is higher.")
elif totval(t1) < totval(t2):
  print("Tuple 2 total value is higher.")
else:
  print("Both tuples have same total value.")
```

# Mission 5

Here's a tuple:

my_tuple = ("data analyst", "data scientist", "data engineer", "data architect")

Create a script that allows you to change one of the values of "my_tuple" by indicating the position of the value you want to change, as well as the new value.

```
# Function to change a tuple
def tuple_modify(t, index, value):

  # Convert to list
  l = list(t)

  # Change necessary list element
  l[index] = value

  # Convert back to tuple
  t = tuple(l)

  # Return tuple, otherwise it's lost
  return t


# Create my_tuple
my_tuple = ("data analyst", "data scientist", "data engineer", "data architect")

# Modify first element
my_tuple = tuple_modify(my_tuple, 0, "expert on camembert")

# Print out the result
print(my_tuple)
```
